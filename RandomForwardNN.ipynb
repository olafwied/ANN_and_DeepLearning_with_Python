```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# how to do forward propagation (with random weights)

D = 2 # dimensionality of input
H = 3 # hidden layer size
C = 3 # number of classes
N = 500 #sample size per class

#gaussian features
X1 = np.random.randn(N, D) + np.array([0, -2])
X2 = np.random.randn(N, D) + np.array([2, 2])
X3 = np.random.randn(N, D) + np.array([-2, 2])
X = np.vstack([X1, X2, X3])
#label data
Y = np.array([0]*N + [1]*N + [2]*N)

plt.scatter(X[:,0], X[:,1], c=Y, s=100, alpha=0.5)
plt.show()
```

```python
# initial weights (random)
W1 = np.random.randn(D, H)
b1 = np.random.randn(H)
W2 = np.random.randn(H, C)
b2 = np.random.randn(C)

def forward(X, W1, b1, W2, b2):
    Z = 1 / (1 + np.exp(-X.dot(W1) - b1)) #value of hidden layer (sigmoid)
    A = Z.dot(W2) + b2
    expA = np.exp(A) 
    Y = expA / expA.sum(axis=1, keepdims=True) #softmax of A
    return Y

def classification_rate(Y, P):
    """computes number of correctly predicted (P=Predictions) 
    labels (Y=true labels) 
    divided by total number of labels"""
    n_correct = 0
    n_total = 0
    for i in range(len(Y)):
        n_total += 1
        if Y[i] == P[i]:
            n_correct += 1
    return float(n_correct) / n_total

P_Y_given_X = forward(X, W1, b1, W2, b2)
P = np.argmax(P_Y_given_X, axis=1) #get class with highest probability

# verify we chose the correct axis
assert(len(P) == len(Y))

print("Classification rate for randomly chosen weights: %f" %classification_rate(Y, P))
```
